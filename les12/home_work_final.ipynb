```python
from sympy import *
```

```python
x = Symbol('x')
fx = sin(x)**2 - cos(x)**2
```

```python
# корни
solveset(fx, x, S.Reals)
```

```python
# интервал возрастания функции
solveset(fx > 0, x, S.Reals)
```

```python
# интервалы убывания функции
solveset(fx < 0, x, S.Reals)
```

```python
from sympy.plotting import plot
plot(fx, (x, -15, 15))
```

```python
derivative_f = fx.diff(x)
print(derivative_f)
```

```python
# Определить промежутки, на котором f > 0
solveset(derivative_f > 0, x, S.Reals)
```

```python
# Определить промежутки, на котором f < 0
solveset(derivative_f < 0, x, S.Reals)
```

```python
# Определить вершину
solveset(derivative_f, x, S.Reals)
```
